```python
from geopy.geocoders import Nominatim
geolocator = Nominatim(user_agent="pollution_map")

# Add latitude and longitude columns
latitudes = []
longitudes = []

for place in carbon['Country']:  # Replace with the column that has place names
    location = geolocator.geocode(place)
    if location:
        latitudes.append(location.latitude)
        longitudes.append(location.longitude)
    else:
        latitudes.append(None)  # If no geocode result
        longitudes.append(None)
```

```python
carbon['latitude'] = latitudes
carbon['longitude'] = longitudes
carbon['latitude'].dtype
```

# Displaying carbon pollutiono

```python
carbon.head(10)
```

```python
# Convert the column to numeric, forcing errors to NaN
carbon['CO2 emissions, latest year'] = pd.to_numeric(carbon['CO2 emissions, latest year'], errors='coerce')

# Now the column is of type float
print(carbon['CO2 emissions, latest year'].dtype)  # Should be float
```

```python
def get_color(pollution_level):
    if pollution_level < 5000.00:
        return 'green'
    elif pollution_level < 10000.00:
        return 'yellow'
    elif pollution_level < 15000.0:
        return 'orange'
    else:
        return 'red'
```

```python
import folium
from folium import plugins

m = folium.Map(
    location=[carbon['latitude'].mean(), carbon['longitude'].mean()], 
    zoom_start=6,
    tiles='CartoDB positron',  
    attr="Map tiles by CartoDB, under CC BY 3.0."
)

for idx, row in carbon.iterrows():
    color = get_color(row['CO2 emissions, latest year'])
    folium.CircleMarker(
        location=[row['latitude'], row['longitude']],
        radius=7,
        color=color,
        fill=True,
        fill_color=color,
        fill_opacity=0.6,
        popup=f"{row['Country']} Carbon Pollution Level: {row['CO2 emissions, latest year']}"
    ).add_to(m)
```

# Displaying energy pollution

```python
energy.head()
```

```python
from geopy.geocoders import Nominatim
geolocator = Nominatim(user_agent="pollution_map")

# Add latitude and longitude columns
latitudes = []
longitudes = []

for place in energy['Country and area']:  # Replace with the column that has place names
    location = geolocator.geocode(place)
    if location:
        latitudes.append(location.latitude)
        longitudes.append(location.longitude)
    else:
        latitudes.append(None)  # If no geocode result
        longitudes.append(None)
```

```python
energy.head(10)
```

```python
energy['latitude'] = latitudes

energy['longitude'] = longitudes
```

```python
energy.dropna(subset=['latitude', 'longitude'], inplace=True)
energy.head(10)
```

```python
def get_color(energy_supply):
    if energy_supply < 50.0:  # Low energy supply (e.g., small or developing countries)
        return 'green'
    elif energy_supply < 200.0:  # Moderate energy supply (e.g., mid-sized countries)
        return 'yellow'
    elif energy_supply < 500.0:  # High energy supply (e.g., larger developed countries)
        return 'orange'
    else:  # Very high energy supply (e.g., major energy producers like the US, China, Russia)
        return 'red'
```

```python
import folium
from folium import plugins

m = folium.Map(
    location=[energy['latitude'].mean(), energy['longitude'].mean()], 
    zoom_start=5, 
    tiles='CartoDB positron',  
    attr="Map tiles by CartoDB, under CC BY 3.0."
)

for idx, row in energy.iterrows():
    color = get_color(row['Energy supply (petajoules)'])
    folium.CircleMarker(
        location=[row['latitude'], row['longitude']],
        radius=7,
        color=color,
        fill=True,
        fill_color=color,
        fill_opacity=0.6,
        popup=f"{row['Country and area']} Energy Pollution Level: {row['Energy supply (petajoules)']}"
    ).add_to(m)
```
